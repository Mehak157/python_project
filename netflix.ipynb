```python
#‚òÅÔ∏è WordCloud Ideas:
"""Most used words in movie titles

Common words in descriptions"""
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
set(STOPWORDS)
```

```python
#Most used words in movie titles
tot_title=' '.join(df["title"])
```

```python
wordc=WordCloud(stopwords=set(STOPWORDS)).generate(tot_title)
plt.imshow(wordc)
plt.title("title")
```

```python
tot_movie=df[df["type"]=="Movie"]
tot_move=tot_movie["title"]
total_mov=' '.join(tot_move)
wc=WordCloud(background_color="white",stopwords=set(STOPWORDS)).generate(total_mov)
plt.imshow(wc)
plt.axis('off')
```

```python
#Common words in descriptions
tot_desc=' '.join(df["description"])
wordc2=WordCloud(stopwords=set(STOPWORDS)).generate(tot_desc)
plt.imshow(wordc2)
plt.title("title")
plt.axis('off')
```
