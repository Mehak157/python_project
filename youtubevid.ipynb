```python
from wordcloud import WordCloud, STOPWORDS
from textblob import TextBlob
import matplotlib.pyplot as plt

# Clean data
df = df[df["Comment"].notnull()]
df["Comment"] = df["Comment"].astype(str)

# Sentiment analysis
df["polarity"] = df["Comment"].apply(lambda x: TextBlob(x).sentiment.polarity)

# Filter positive comments
positive_com = df["polarity"] > 0
total_com = df[positive_com]
total_comment = ' '.join(total_com["Comment"])

# Generate word cloud only if there's data
if total_comment.strip():
    wordcloud = WordCloud(stopwords=set(STOPWORDS)).generate(total_comment)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
else:
    print("‚ö†Ô∏è No positive comments available to generate a word cloud.")
```
