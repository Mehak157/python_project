```python
#1. Ride Volume Over Time
date_r=df["Date"].value_counts().sort_index()
#book_id=df["Booking ID"].value_counts()
plt.plot(date_r.index,date_r.values)
plt.title("Ride Volume Over Time")
plt.xticks(rotation=90)
plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"])  # Ensure it's datetime
ride_counts_by_date = df["Date"].value_counts().sort_index()

plt.plot(ride_counts_by_date.index, ride_counts_by_date.values)
plt.title("Ride Volume Over Time")
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Number of Rides")
plt.tight_layout()
plt.show()
```

```python
#2. Booking Status Breakdown #Booking Status'
book_stats=df["Booking Status"].value_counts().sort_index()
plt.figure(figsize=(10,7))
plt.pie(book_stats,labels=book_stats.index, autopct="%1.1f%%")
plt.show()
```

```python
#3. Top 5 Vehicle Types by Ride Distance #group by,#Vehicle Type#Ride Distance #seaborn#
v_type=df.groupby("Vehicle Type")["Ride Distance"].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='Vehicle Type',y='Ride Distance',data=v_type)
plt.show()
```

```python
vtype=df["Vehicle Type"].value_counts()
#rdist=df["Ride Distance").sort_index()
plt.bar(vtype.index,vtype.values ,data=vtype )
plt.title("top 5 vehicle")
plt.show()
```

```python
vtype=df["Vehicle Type"].value_counts()
rdist=df["Ride Distance"].sort_index()
plt.plot(vtype.index,vtype.values)
plt.title("top 5 vehicle")
plt.show()
```

```python
#4. Average Customer Ratings by Vehicle Type   #Customer Rating  #Vehicle Type
cust_r=df.groupby("Vehicle Type")["Customer Rating"].mean().sort_values(ascending=False)

#typer=df["Vehicle Type"].value_counts()
plt.bar(cust_r.index,cust_r.values)
plt.title("customer ratings")
plt.xticks(rotation=90)

plt.show()
```

```python
#5. cancelled Rides Reasons #Cancelled Rides by Customer#Reason for cancelling by Customer
cancel_rides=df["Reason for cancelling by Customer"].value_counts().sort_index()
plt.figure(figsize=(7,7))
plt.pie(cancel_rides,labels=cancel_rides.index,autopct="%1.1f%%")
plt.show()
```

```python
#6.revenue by Payment Method
payment=["cash","upi"," credit card","paytm","phonepe","debit card"]
value=[23,43,45,43,12,3]
plt.bar(payment,value)
plt.title("Payment method")
plt.show()
```

```python
#7. Top 5 Customers by Total Booking Value #Customer ID #Booking Value
cust_5=(
    df.groupby("Customer ID")["Booking Value"]
.sum()
.sort_values(ascending=False)
.head(10)
.reset_index()
)
sns.barplot(x='Customer ID',y='Booking Value',data=cust_5 )
plt.xticks(rotation=90)
plt.show()
```

```python
cust_5 = (
    df.groupby("Customer ID")["Booking Value"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()  #  Now it's a DataFrame with 'Customer ID' and 'Booking Value' columns
)
sns.barplot(x="Customer ID", y="Booking Value", data=cust_5, palette="plasma")
plt.title("Top 10 Customers by Total Booking Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#8. Ride Distance Distribution Per Day
 #'Ride Distance'
#Vehicle Type
ride=df["Ride Distance"]
plt.figure(figsize=(10,7))
plt.hist(ride,bins=20, edgecolor="black")
plt.show()
```

```python
ride=df.groupby("Ride Distance")["Vehicle Type"].value_counts().reset_index(name="count")
#typev=df["Vehicle Type"]
ax=sns.barplot(data=ride,x="Vehicle Type",y="Ride Distance")
plt.xticks(rotation=90)
plt.show()


```

```python
#9. Driver Ratings Distribution
rate_d=df.groupby("Vehicle Type")["Driver Ratings"].mean().sort_index().reset_index()
sns.barplot(data=rate_d, x="Vehicle Type",y="Driver Ratings")
plt.title("Driver Ratings")
plt.show()
```

```python
#10. Customer vs. Driver Ratings# Driver Ratings#Customer Rating#"Vehicle Type"
data_d=df.groupby("Vehicle Type")[["Driver Ratings","Customer Rating"]].mean().sort_index().reset_index()
data_2=pd.melt( data_d,
               id_vars="Vehicle Type",
               value_vars=["Driver Ratings","Customer Rating"],
               var_name="Rating type",
               value_name="Avg ratings")
sns.lineplot(data=data_2 , x="Vehicle Type" , y="Avg ratings",hue="Rating type",marker="o")
plt.show()
```

```python
# Group and prepare data
data_d = df.groupby("Vehicle Type")[["Driver Ratings", "Customer Rating"]].mean().reset_index()

# Melt the data into long format
long_df = pd.melt(data_d, id_vars="Vehicle Type", 
                  value_vars=["Driver Ratings", "Customer Rating"], 
                  var_name="Rating Type", value_name="Average Rating")

# Plot with two lines
plt.figure(figsize=(10, 6))
sns.lineplot(data=long_df, x="Vehicle Type", y="Average Rating", hue="Rating Type", marker="o")
plt.title("Driver vs Customer Ratings by Vehicle Type")
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.xlabel("Vehicle Type")
plt.tight_layout()
plt.show()
```
