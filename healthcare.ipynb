```python
bed_c=df1["Bed_Occupancy"].value_counts().sort_index()
plt.bar(bed_c.index , bed_c.values)
plt.title("bed occupacy")
plt.show()
```

```python
#feedback by doctor #Feedback
feed_d=df1["Feedback"].value_counts().reset_index().sum()
plt.pie(feed_d,labels=feed_d.values , autopct="%1.1f%%")
#plt.pie(feed_d,labels=feed_d.index, autopct="%1.1f%%")
plt.title("feedback by doctor")
plt.show()
```

```python
bill_in=df1.groupby("Diagnosis")[["Billing Amount","Health Insurance Amount"]].mean().sort_index().reset_index()
bill_init=pd.melt(bill_in,
    id_vars="Diagnosis",
    value_vars=["Billing Amount","Health Insurance Amount"],
    var_name="Bill amount",
     value_name="insurence bill"
               )
sns.lineplot( data=bill_init, x="Diagnosis" ,y="insurence bill" , hue="Bill amount" )
plt.title("Bill Amount")
plt.show()
```

```python
deses=df1["Diagnosis"].value_counts().sort_index(ascending=False)
plt.bar(deses.index,deses.values)
plt.title("Diagnosis")
plt.show()
```
