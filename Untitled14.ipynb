```python
#Generate separate WordClouds of subjects chosen by Male and Female students.
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Filter data for Male and Female students
male_subjects = df[df["Gender"] == "Male"]["Subject"].astype(str)
female_subjects = df[df["Gender"] == "Female"]["Subject"].astype(str)

# Join all subject names into single strings
male_subject_text = ' '.join(male_subjects)
female_subject_text = ' '.join(female_subjects)

# Generate wordclouds
male_wc = WordCloud(width=800, height=400, background_color='white').generate(male_subject_text)
female_wc = WordCloud(width=800, height=400, background_color='white').generate(female_subject_text)

# Plot both side by side
plt.figure(figsize=(16, 6))

# Male WordCloud
plt.subplot(1, 2, 1)
plt.imshow(male_wc, interpolation='bilinear')
plt.title("Subjects Chosen by Male Students")
plt.axis('off')

# Female WordCloud
plt.subplot(1, 2, 2)
plt.imshow(female_wc, interpolation='bilinear')
plt.title("Subjects Chosen by Female Students")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
"""#Create WordClouds of students who received:

Grade A

Grade D
and compare the word clouds."""
stude_A=df[df["Grade"]=="A"]["Student"].astype(str)
stude_B=df[df["Grade"]=="B"]["Student"].astype(str)
stude_C=df[df["Grade"]=="C"]["Student"].astype(str)

joinA=''.join(stude_A)
joinB=''.join(stude_B)
joinC=''.join(stude_C)

A_cloud=WordCloud(width=800, height=400, background_color='white').generate(joinA)
B_cloud=WordCloud(width=800, height=400, background_color='white').generate(joinB)
C_cloud=WordCloud(width=800, height=400, background_color='white').generate(joinC)

plt.figure(figsize=(16,6))
#plot subplot
plt.subplot(1,3,1)
plt.imshow(A_cloud,interpolation='bilinear')
plt.title("Grade A")
plt.axis('off')
#plot subplot2
plt.subplot(1,3,2)
plt.imshow(B_cloud,interpolation='bilinear')
plt.title("Grade B")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(B_cloud,interpolation='bilinear')
plt.title("Grade C")
plt.axis('off')
```
